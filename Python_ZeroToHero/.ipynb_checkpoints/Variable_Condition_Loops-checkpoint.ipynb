```python
condition = True
itr = 1
while condition:
    if(itr==5):
        print("Continue-----")
        continue
    if itr > 11:
        print("Stop Iteration")
        break
    print("Loop Condition: %s" %condition + " Iteration %i" %itr)
    itr+=1
```
